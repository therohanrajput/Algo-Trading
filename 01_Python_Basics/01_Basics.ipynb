### <span style= "color:#54b8e3;"> 14. Break-Even Price Calculator
    Objective: Calculate the break-even price after buying a stock and spending money on fees.

```python
stock_price = 100.00
stock_quantity = 10
fees = 50.00

total_cost = (stock_price * stock_quantity) + fees
break_even_price = total_cost / stock_quantity

print(f"Total cost of the trade is: ${total_cost:.2f}")
print(f"Break-even price is: ${break_even_price:.2f}")
```

### <span style="color: #54b8e3;">16. Calculate Required Margin
    Objective: Calculate the required margin for a trade based on user inputs for trade size and margin rate.

```python
stock_price = 200.00
lot_size = 10
fee_per_stock = 0.02
margin_requirement = 0.25
##Total fees for the trade
total_fees = stock_price * lot_size * fee_per_stock
##Total cost of the trade including fees
total_cost = (stock_price * lot_size) + total_fees
##Total cost of the trade without fees
total_cost_without_fees = stock_price * lot_size
##Margin required for the trade
margin_required = total_cost_without_fees * margin_requirement

print(f"Total fees for the trade are: ${total_fees:.2f}")
print(f"Total cost of the trade including fees is: ${total_cost:.2f}")
print(f"Margin required for the trade is: ${margin_required:.2f}")
```

### <span style="color: #54b8e3;">18. Calculate the Weighted Average Price
    Objective: Calculate the weighted average price based on user inputs for different prices and their respective weights.

```python
stock_price_1 = 150.00
stock_quantity_1 = 20
stock_price_2 = 250.00
stock_quantity_2 = 10

total_cost_stock_1 = stock_price_1 * stock_quantity_1
total_cost_stock_2 = stock_price_2 * stock_quantity_2
total_quantity = stock_quantity_1 + stock_quantity_2

average_weighted_price = (total_cost_stock_1 + total_cost_stock_2) / total_quantity
print(f"Average weighted price of the two stocks is: ${average_weighted_price:.2f}")
```
