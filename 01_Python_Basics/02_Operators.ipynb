### Objective 1 :-  Calculate potential profit, risk, and other financial metrics for a day trade based on given stock price, account balance, and predefined trading rules.

```python
entry_price = 150.0  # Entry price of the stock
exit_price = 155.0   # Exit price of the stock
stop_loss_price = 148.0  # Stop loss price of the stock
share_count = 10  # Number of shares bought
account_balance = 2000.0  # Initial account balance

# Calculate potential profit
potential_profit = (exit_price - entry_price) * share_count

# Calculate potential loss
potential_loss = (entry_price - stop_loss_price) * share_count

# Calculate risk-reward ratio
if potential_loss != 0:
    risk_reward_ratio = potential_profit / potential_loss
else:
    risk_reward_ratio = float('inf')

# Calculate new account balance after the trade
new_account_balance = account_balance + potential_profit - potential_loss

print("Potential Profit:", potential_profit)
print("Potential Loss:", potential_loss)
print("Risk Reward Ratio:", risk_reward_ratio)
print("New Account Balance:", new_account_balance)
```

### 2. Initial price and Volatility

```python
import random
import datetime
initial_price = 150.0  # Initial price of the stock
volatility = 0.02  # Volatility of the stock (2%)

volatility_change = random.choice([-volatility, volatility])  # Randomly choose to increase or decrease the price

daily_price_change = initial_price +(initial_price * volatility_change)  # Calculate the daily price change

today = datetime.datetime.today()  # Get today's date

print("Date:", today)
print("Initial Price:", initial_price)
print("Daily Price Change:", daily_price_change)
```
