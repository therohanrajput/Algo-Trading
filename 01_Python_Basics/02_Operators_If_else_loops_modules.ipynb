### 1. Calculate potential profit, risk, and other financial metrics for a day trade based on given stock price, account balance, and predefined trading rules.

```python
entry_price = 150.0  # Entry price of the stock
exit_price = 155.0   # Exit price of the stock
stop_loss_price = 148.0  # Stop loss price of the stock
share_count = 10  # Number of shares bought
account_balance = 2000.0  # Initial account balance

# Calculate potential profit
potential_profit = (exit_price - entry_price) * share_count

# Calculate potential loss
potential_loss = (entry_price - stop_loss_price) * share_count

# Calculate Reward-to-Risk Ratio (colloquially "Risk-Reward Ratio")
if potential_loss != 0:
    reward_risk_ratio = potential_profit / potential_loss
else:
    reward_risk_ratio = float('inf')

# Calculate mathematical Risk-to-Reward Ratio
if potential_profit != 0:
    risk_reward_ratio = potential_loss / potential_profit
else:
    risk_reward_ratio = float('inf')

# Calculate percentage of account balance risked on this trade
account_risk_percent = (potential_loss / account_balance) * 100

#Calculate percentage of account balance profited on this trade
account_profit_percent = (potential_profit / account_balance) * 100

# Calculate new account balances based on the mutually exclusive outcomes
new_balance_if_won = account_balance + potential_profit
new_balance_if_lost = account_balance - potential_loss

print(f"Potential Profit: ${potential_profit:.2f}")
print(f"Potential Loss: ${potential_loss:.2f}")
print(f"Reward-to-Risk Ratio (Reward : Risk): {reward_risk_ratio:.2f} : 1")
print(f"Risk-to-Reward Ratio (Risk : Reward): {risk_reward_ratio:.2f} : 1")
print(f"Account Risk: {account_risk_percent:.2f}%")
print(f"Account Profit: {account_profit_percent:.2f}%")
print(f"New Balance if trade succeeds (won): ${new_balance_if_won:.2f}")
print(f"New Balance if stop loss hit (lost): ${new_balance_if_lost:.2f}")
```

### 2. Initial price and Volatility

```python
import random
import datetime
initial_price = 150.0  # Initial price of the stock
volatility = 0.02  # Volatility of the stock (2%)

volatility_change = random.choice([-volatility, volatility])  # Randomly choose to increase or decrease the price

daily_price_change = initial_price +(initial_price * volatility_change)  # Calculate the daily price change

today = datetime.datetime.today()  # Get today's date

print("Date:", today)
print("Initial Price:", initial_price)
print("Daily Price Change:", daily_price_change)
```

### 3. You have a list of stock prices over 30 days. Write a logical expression that will identify consecutive price drops over three days or more. For instance, `[120, 118, 115, 110]` would qualify, but `[120, 118, 120, 110]` would not.

```python
prices = [123,118,115]
number_of_drops = 3
if len(prices) < number_of_drops + 1:
      # Not enough data to determine consecutive drops
    consecutive_drops = False
else:
    consecutive_drops = any(prices[i] > prices[i + 1] >prices[i + 2]> prices[i + 3] for i in range(len(prices)-number_of_drops))
if consecutive_drops:
    print(f"There are at least {number_of_drops} consecutive drops in the price data.")
else:    
    print(f"There are not {number_of_drops} consecutive drops in the price data.")
```

### 4. You are designing an algorithm to trade stocks. If the current price of a stock is above the 30-day moving average, print `"Buy Signal"`; otherwise, print `"Sell Signal"`. Assume you have a list of daily prices and that the moving average can be calculated.

```python
price_list_30 = [150, 148, 149, 151, 152, 153, 154, 155, 156, 157, 158, 159, 160, 161, 162, 163, 164, 165, 166, 167, 168, 169, 170, 171, 172, 173, 174, 175, 150]
moving_average_30 = sum(price_list_30) / (len(price_list_30))
current_price = price_list_30[-1]  # Assuming the current price is the last price in the list

print(f"30-day Moving Average: {moving_average_30:.2f}")
print(f"Current Price: {current_price:.2f}")

if current_price > moving_average_30:
    print("Buy Signal")
else:
    print("Sell Signal")
```
