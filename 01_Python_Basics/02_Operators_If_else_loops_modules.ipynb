### <span style="color: #54b8e3;">1. Calculate potential profit, risk, and other financial metrics for a day trade based on given stock price, account balance, and predefined trading rules.

```python
entry_price = 150.0  # Entry price of the stock
exit_price = 155.0   # Exit price of the stock
stop_loss_price = 148.0  # Stop loss price of the stock
share_count = 10  # Number of shares bought
account_balance = 2000.0  # Initial account balance

# Calculate potential profit
potential_profit = (exit_price - entry_price) * share_count

# Calculate potential loss
potential_loss = (entry_price - stop_loss_price) * share_count

# Calculate Reward-to-Risk Ratio (colloquially "Risk-Reward Ratio")
if potential_loss != 0:
    reward_risk_ratio = potential_profit / potential_loss
else:
    reward_risk_ratio = float('inf')

# Calculate mathematical Risk-to-Reward Ratio
if potential_profit != 0:
    risk_reward_ratio = potential_loss / potential_profit
else:
    risk_reward_ratio = float('inf')

# Calculate percentage of account balance risked on this trade
account_risk_percent = (potential_loss / account_balance) * 100

#Calculate percentage of account balance profited on this trade
account_profit_percent = (potential_profit / account_balance) * 100

# Calculate new account balances based on the mutually exclusive outcomes
new_balance_if_won = account_balance + potential_profit
new_balance_if_lost = account_balance - potential_loss

print(f"Potential Profit: ${potential_profit:.2f}")
print(f"Potential Loss: ${potential_loss:.2f}")
print(f"Reward-to-Risk Ratio (Reward : Risk): {reward_risk_ratio:.2f} : 1")
print(f"Risk-to-Reward Ratio (Risk : Reward): {risk_reward_ratio:.2f} : 1")
print(f"Account Risk: {account_risk_percent:.2f}%")
print(f"Account Profit: {account_profit_percent:.2f}%")
print(f"New Balance if trade succeeds (won): ${new_balance_if_won:.2f}")
print(f"New Balance if stop loss hit (lost): ${new_balance_if_lost:.2f}")
```

   ### <span style="color: #54b8e3;">7. (for loop) You have data on the opening and closing prices of multiple stocks in a portfolio. Using a for loop, calculate and print the total daily gain or loss for each stock in the portfolio. 

```python
import random #to grenrate random prices and choice

#ticker list of popular stock
ticker_list = ["AAPL", "MSFT", "GOOGL", "AMZN", "NVDA", "TATA", "BANKNIFTY"]

#Empty list where randomly genrate portfolio will be stored
prortfolio =[]
#Generate 3 Unique stocks data without repeating them via for loop and random.sample
for ticker in random.sample(ticker_list, 3):

    #Generate a random openning price between $50 to $500 to 2 decimal point via round function
    open_price = round(random.uniform(50.0,500.0),2)

    #Generate a realstick change percent within range of only +5 and -5 %
    #prevent closing price from being completely unrealstick
    percent_change = random.uniform(0.05, -0.05)

    #Generate a realstic closing price for the stock
    close_price = round(open_price * (1+ percent_change),2)

    #bunddle this all data into the directory
    stock_directory = {
        "ticker": ticker,
        "open": open_price,
        "close": close_price
    }
    #append created stock directory to protfolio list
    prortfolio.append(stock_directory)
#print portfolio to check is list is created right or not
#print(prortfolio)

# ----------------- CALCULATE GAIN/LOSS -----------------

print("-------Generated Portfolios & Daily Performance--------")

for stock in prortfolio:
    ticker = stock["ticker"]
    open_price = stock["open"]
    close_price = stock["close"]

    #Calculate daily gain loss
    daily_gain_loss = close_price - open_price

    if open_price > close_price:
        print(f"{ticker}: (Opened : ${open_price} | Closed : ${close_price}) -> Total Loss : ${daily_gain_loss:.2f}:")
    elif open_price < close_price:
        print(f"{ticker}: (Opened : ${open_price} | Closed : ${close_price}) -> Total Gain : ${daily_gain_loss:.2f}")
    else:
        print(f"{ticker}: (Opened : ${open_price} | Closed : ${close_price}) -> No Change ($0.00).")
```

### <span style="color: #54b8e3;">9. You have a complex trading rule where a `"Strong Buy"` signal is given if the stock price is above both the 30-day and 50-day moving averages, a `"Buy"` signal is given if only above the 30-day, a `"Hold"` signal if above the 50-day, and a `"Sell"` signal otherwise. Calculate the signals assuming you have daily prices for at least 50 days.

```python
import random

#empty list for prices range atleast 60 days


#config our price generation parameters
starting_price = 100.0 #stock price on day 1
total_days = 60 #total no of day for we want to simulate the price
prices = [starting_price] #initialize prices list with the starting price

for price in range(total_days-1):
    #print(price) 

    #generating daily return between 2.5% to -2.5% so look realstick price change
    daily_return = random.uniform(0.025, -0.025)

    #calculate new price based on the previous day price
    new_price = round(prices[-1] * (1+ daily_return), 2)
    prices.append(new_price)

#print(prices)
#print(len(prices))

print(f"--- Successfully Generated {total_days} Days of Random Stock Prices ---")
#print("First 5 Days:", prices[:5])
print("Last 10 Days:", prices[-10:])
print("\n--- Trading Signals (From Day 50 onwards) ---")

for i in range(49, len(prices)):
    current_price = prices[i]
    #first calculate 30 moving average from (i-29 to i)
    ma30 = sum(prices[i-29 : i+1])/30

    # calculate 50 moving average from (i-29 to i)
    ma50 = sum(prices[i-49 : i+1])/50

    if current_price > ma30 and current_price >ma50:
        signal = "Strong Buy"
    elif current_price > ma30:
        signal = "Buy"
    elif current_price > ma50:
        signal = "Hold"
    else:
        signal = "Sell"

    print(f"Day {i+1} | Price: ${current_price} | MA30: ${ma30} | MA50: ${ma50} | Signal: {signal}")
```

### <span style="color: #54b8e3;">11. Calculate the best day to buy and the best day to sell to maximize profit within a list of daily prices using a single pass of nested loops.

```python
#define the prices list
prices = [100,98,110,101,85,88,89,99]

#initialize neccessary variables with 0
max_profit = 0
best_buy = 0
best_sell = 0

#outer loop to calculate the best buy
for i in range(len(prices)):
    for j in range(i+1, len(prices)):
        #first calculate the profit
        profit = prices[j] - prices[i]
        #check condition and update variable accordingly
        if profit > max_profit:
            max_profit = profit #update max profit
            best_buy = i #update best buy day
            best_sell = j #update best sell day

print("---------Best Day to Buy/Sell----------")
print(f"Best Day to buy: {best_buy+1} | Price: ${prices[best_buy]}")
print(f"Best Day to Sell: {best_sell+1} | Price: ${prices[best_sell]}")
print(f"Maximum Profit: ${max_profit}")
```
