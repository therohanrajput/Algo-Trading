### <span style="color: #54b8e3;">1. **Complex String Manipulation with Loops**  
   - Given a string of mixed data containing symbols and prices like `"AAPL100GOOG200MSFT300"`, write a script to extract and print each symbol with its corresponding price on separate lines.

```python
from numpy._core import fromnumeric
def extract_symbol_and_price(data_string):
    #list to store extracted symbol and price
    pairs = []
    #string buffer to store current symbol and price
    current_symbol= ""
    current_price = ""
    #loop throught each character in the input string
    for char in data_string:
        #if character is alphabetic then they belongs to the stock symbol
        if char.isalpha():
            if current_price:
                #Inside the alphabetic check, if current_price is not empty, it means we were previously building a price (e.g., "100") and have now hit the first character of the next stock symbol (e.g., 'G' from GOOG). This marks the transition between stock entries.
                pairs.append((current_symbol,current_price))
                current_symbol = ""
                current_price =""
            ###Append the alphabetic character to the current symbol buffer
            current_symbol += char 
                       
        elif char.isdigit():
            #If the character is not alphabetic, checks if it is a digit (0-9).
            if current_symbol:
                #Appends the digit to the current_price buffer.
                current_price += char
    if current_symbol and current_price:
        pairs.append((current_symbol, current_price))
        
    # Return the collected pairs
    return pairs



Str = "AAPL100GOOG200MSFT300"
results = extract_symbol_and_price(Str)

for symbol, price in results:
    print(f"Symbol: {symbol}, Price: {price}")
    



```

### <span style="color: #54b8e3;"> 3. String Slicing and Indexing
   - Given a long string of security transaction details formatted as "Buy100AAPLSell200GOOG", where each segment starts with a transaction type followed by the amount and the ticker, print each transaction detail on a new line.

```python
def parse_transactions(transaction_data):
    #Identify the each transtion segment by locationg buy and sell
    segment_indices = []
    i=0
    while i < len(transaction_data):
        if transaction_data[i:i+3] == 'Buy':
            segment_indices.append(i)
            i += 3
        elif transaction_data[i:i+4] == 'Sell':
            segment_indices.append(i)
            i += 4
        else:
            i += 1
    #now append the total string len to the segent_indices 
    #it define the end boundary of the last segment to slice string cleanly
    segment_indices.append(len(transaction_data))

    #now process each segment using slicing
    for indx in range(len(segment_indices) - 1):
        start = segment_indices[indx]  # start = 0
        end = segment_indices[indx+1]  # end = 10
        # Slice out the individual segment (e.g., "Buy100AAPL")
        segment = transaction_data[start:end]
        
        # Extract the Transaction Type using a slice
        if segment.startswith("Buy"):
            tx_type = segment[0:3]  # "Buy" is 3 characters
            remainder = segment[3:]
        else:
            tx_type = segment[0:4]  # "Sell" is 4 characters
            remainder = segment[4:]
            
        # Locate where the digit amount ends and the ticker symbol begins
        # We find the index of the first alphabetic character in the remainder
        ticker_start_idx = 0
        for char_idx, char in enumerate(remainder):
            if char.isalpha():
                ticker_start_idx = char_idx
                break
                
        # 6. Slice the remainder to separate the Amount and the Ticker
        amount = remainder[:ticker_start_idx]
        ticker = remainder[ticker_start_idx:]
        
        # 7. Print the extracted details on a new line
        print(f"Action: {tx_type:<4} | Qty: {amount:<4} | Ticker: {ticker}")


string_transaction = "Buy100AAPLSell200GOOG"
parse_transactions(string_transaction)

```
